```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
# Read csv file into a pandas dataframe and check first 10 lines
df = pd.read_csv("miss_reg.csv", header = 0)
df.head(50)
```

```python
# Identify the feature with missing values and the predictors
target_column = 'Income'
predictor_columns = ['Age']
```

```python
# Split data into complete and missing parts
df_complete = df.dropna()
df_missing = df[df[target_column].isnull()]
```

```python
# Prepare data for regression
X_train = df_complete[predictor_columns].values
y_train = df_complete[target_column].values
X_test = df_missing[predictor_columns].values
```

```python
# Fit the regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict missing values
imputed_values = model.predict(X_test)
```

```python
# Fill in missing values
df.loc[df[target_column].isnull(), target_column] = imputed_values
```
