```python
import pandas as pd
import numpy as np
df = pd.read_csv("student.csv", header = 0)
df.head()
```

```python
#Create data seat with filters
filter=df.Employed=='yes'
df1=df[filter]
df1.head(10)
```

```python
#Find the categorical column and separate it out with a different dataframe. 
#To do so, use the select_dtypes() function from pandas:
df_categorical = df.select_dtypes(exclude=[np.number])
df_categorical
```

```python
#Find the distinct unique values in the Grade column. 
#To do so, use the unique() function from pandas with the column name:
df_categorical['Grade'].unique()
```

```python
#For the Gender column, write the following code:
df_categorical.Gender.value_counts()
```

```python
#Replace the entries in the Grade column,ignor error if any. 
#Replace 1st class with 1, 2nd class with 2, and 3rd class with 3. To do so, use the replace() function:
df_categorical.Grade.replace({"1st Class":1, "2nd Class":2, "3rd Class":3}, inplace= True)
```

```python
df_categorical
```

```python
#Replace the entries in the Gender column. 
#Replace Male with 0 and Female with 1. To do so, use the replace() function:
df_categorical.Gender.replace({"Male":0,"Female":1}, inplace= True)
#Replace no with 0 and yes with 1. To do so, use the replace() function:
df_categorical.Employed.replace({"yes":1,"no":0}, inplace = True)
```
