```python
# Making a list of missing value types
missing_values = ["n/a", "na", "--"]
# Load data from CSV file
df = pd.read_csv("housing_students.csv", na_values = missing_values)
```

```python
# Display first few rows of the dataset
df.head(10)
```

```python
# Replace using mean
mean = df['housing_median_age'].mean()
# Fill missing values with specified value
df['housing_median_age'].fillna(mean, inplace=True)
df['housing_median_age']
```

```python
# Replace using median 
median = df['total_rooms'].median()
# Fill missing values with specified value
df['total_rooms'].fillna(median, inplace=True)
df['total_rooms']
```

```python
# Replace using constant
df['population'].fillna(1000, inplace=True)
df['population']
```

```python
df2 = df.interpolate(method ='linear', limit_direction ='forward')
```

```python
# Display first few rows of the dataset
df2.head(10)
```

```python
# Display column names
for col in df2.columns:
    df2[col] = df2[col].round(1)
```

```python
# Export data to CSV file
df2.to_csv(r'filled_student_housing.csv', index=False)
```
